```python
# Save the downloaded records into an array for easier processing
passwords = []

lines = [line.rstrip('\n') for line in open('100_all.csv')]
lines_1 = [line.rstrip('\n') for line in open('100-1.csv')]
```

### Extract the Passwords

```python

import json

for line in lines:
    j = json.loads(line)
    results = j['results']
    for result in results:
        passwords.append(result['login']['password'])
```

```python
print(passwords[0:10])
```

```python
from zxcvbn import zxcvbn
import json

# results = zxcvbn(pwd)
pwd_scores = []
for pwd in passwords:
    #     print(zxcvbn(pwd)['score'])
    pwd_scores.append(zxcvbn(pwd)['score'])
    #online_no_throttling_10_per_second
#     print(zxcvbn(pwd))
```

```python
print(len(pwd_scores))
```

```python
print(pwd_scores[0:10])
```

### Build the scoring list

```python
pwd_score_mapping = {}
for score in pwd_scores:
    if score in pwd_score_mapping.keys():
        pwd_score_mapping[score]+=1
    else:
        pwd_score_mapping[score] = 1
        
        
print(pwd_score_mapping)

score_sum = 0
for k in pwd_score_mapping.keys():
    score_sum+=pwd_score_mapping[k]

for k in pwd_score_mapping.keys():
    print(pwd_score_mapping[k] * 1.0 / score_sum)


print(score_sum)
```

### Analyze the years distribution

```python
years_count = {}

for result in all_results:
    byear = result['dob'].split("-")[0]
    if byear in years_count.keys():
        years_count[byear] += 1
    else:
        years_count[byear] = 1

print(years_count)
```
